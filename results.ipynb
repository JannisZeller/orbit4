# Assessing Simulation Results with Python

### Setup and Data

---

```python
import numpy as np
import pandas as pd
from core import plot as pl

system_name = "earth_moon_sun" # choice of [earth_mars_sun, earth_moon_sun, tatooine]
animate = False
```

```python
data = pd.read_csv("./data/data_" + system_name + ".csv", sep=",", header=None)
system = pl.System(data)
```

### Plotting General Dynamics

---

```python
_= system.plot_history(figsize=(7, 7), setmax=False)
```

### Plotting Selected Coordinates

---

```python
## Extracting np-values form data
X = system.reshape_data(data.values[:, :6], 2)
X = X[0,:] - X[1, :]
_ = system.plot_custom(X, figsize=(7, 7))
```
